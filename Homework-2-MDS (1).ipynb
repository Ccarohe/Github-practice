## Question 3
Which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data? If there are multiple players who have tied for the best percentage, then sort on the column with player names and submit your answer as a pd.DataFrame object.

```python

df_2000 = df[df['Year'] >= 2000]
max_ft_percentage = df_2000['FT%'].max()

best_players = df_2000[df_2000['FT%'] == max_ft_percentage]#selecting only rows
best_players_sorted = best_players.sort_values(by='Player')

result_df = pd.DataFrame(best_players_sorted)
print(result_df)
```

## Question 4
Among those with at least 50 attempts ('FTA'), which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data?

```python

df_50 = df_2000[df_2000['FTA'] >= 50]
max_50 = df_50['FT%'].max()

best_players = df_50[df_50['FT%'] == max_50]
best_players_50 = best_players.sort_values(by='Player')

result_50 = pd.DataFrame(best_players_50)

print(result_50)
```

## Question 12
Now return the data from #11 back into a long (tidy) format by moving Team (TM), Points and Year into their own columns (Hint: Your data shape should be 90 x 3).

```python
subset_1995_1997 = df[(df['Year'] >= 1995) & (df['Year'] <= 1997)]

overall_average_points= subset_1995_1997.groupby('Tm')['PTS'].mean().sort_values(ascending=False).reset_index(name='Average Points')

average_points_final = overall_average_points[['Average Points']]

average_points_final.shape
```

```python
### 11.Reshape this data in the previous question into a wide format. Create a wide dataset that keeps year in a single row, but spreads team names to multiple columns with each column delineating points per team in 1995-1997 (Hint: Your final answer should have a shape of 3 x 30).
average_points_by_year_team = subset_1995_1997.groupby(['Year', 'Tm'])['PTS'].mean().reset_index()

wide_data = average_points_by_year_team.pivot(index='Year', columns='Tm', values='PTS')

wide_data.shape
wide_data.head()
```

```python
### 12.Now return the data from #11 back into a long (tidy) format by moving Team (TM), Points and Year into their own columns (Hint: Your data shape should be 90 x 3).
long_data = wide_data.reset_index().melt(id_vars=['Year'], var_name='Tm', value_name='PTS')
long_data.shape
```
